# FOLIUM

untuk membuat beberapa jenis peta leaflet, untuk melihat tempat kejadian kejahatan dengan menampilkan peta.

```python
#!conda install -c conda-forge folium=0.5.0 --yes
!pip install folium
import folium

print('Folium installed and imported!')
```

```python
df_crime = pd.read_csv('city_of_london_street.csv')
```

```python
latitude = 51.518494
longitude = -0.111962
```

```python
london_map = folium.Map(location=[latitude, longitude], zoom_start=12)

london_map
```

```python
incidents = folium.map.FeatureGroup()
for lat, lng, in zip(df_crime.Latitude, df_crime.Longitude):
    incidents.add_child(
        folium.CircleMarker(
            [lat, lng],
            radius=5,
            color='yellow',
            fill=True,
            fill_color='blue',
            fill_opacity=0.6
        )
    )

london_map.add_child(incidents)
```

```python
crime = folium.map.FeatureGroup()
for lat, lng, in zip(df_crime.Latitude, df_crime.Longitude):
    crime.add_child(
        folium.CircleMarker(
            [lat, lng],
            radius=5, # define how big you want the circle markers to be
            color='yellow',
            fill=True,
            fill_color='blue',
            fill_opacity=0.6
        )
    )

# add pop-up text to each marker on the map
latitudes = list(df_crime.Latitude)
longitudes = list(df_crime.Longitude)
labels = list(df_crime.Crime_type)

for lat, lng, label in zip(latitudes, longitudes, labels):
    folium.Marker([lat, lng], popup=label).add_to(london_map)    
    
# add incidents to map
london_map.add_child(crime)
```

```python
london_map = folium.Map(location=[latitude, longitude], zoom_start=12)

for lat, lng, label in zip(df_crime.Latitude, df_crime.Longitude, df_crime.Crime_type):
    folium.CircleMarker(
        [lat, lng],
        radius=5, 
        color='yellow',
        fill=True,
        popup=label,
        fill_color='blue',
        fill_opacity=0.6
    ).add_to(london_map)

london_map
```

```python
from folium import plugins

london_map = folium.Map(location = [latitude, longitude], zoom_start = 12)

crime = plugins.MarkerCluster().add_to(london_map)

# loop through the dataframe and add each data point to the mark cluster
for lat, lng, label, in zip(df_crime.Latitude, df_crime.Longitude, df_crime.Crime_type):
    folium.Marker(
        location=[lat, lng],
        icon=None,
        popup=label,
    ).add_to(crime)

# display map
london_map
```
